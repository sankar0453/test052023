#### 6. What happens to variables in a local scope when the function call returns?

```python
def myfunc():
    loc_var=100 # Local variable within the function
    print(loc_var)

myfunc() #100
print(loc_var) #NameError: name 'loc_var' is not defined

"""if we try to access the variable x outside the function, after the function call has returned, 
it will raise a NameError because the variable x was defined within the local scope of the function
and is no longer accessible."""
```

#### 9. How do you make a function variable refer to the global variable?

```python
x=10
def my_funcation():
    global x
    x=20

print(x)  # Output: 10
my_funcation()
print(x) # Output: 20
```
